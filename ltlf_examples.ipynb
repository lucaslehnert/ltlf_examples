## LTLf to DFA Conversion Examples

This notebook lists how two example expressions, "G a" and "F a", are converted into Buechi automata by [Spot](https://spot.lrde.epita.fr/) and DFA by [LTLfKit](https://bitbucket.org/acamacho/ltlfkit/src/master/). The cells below print the HOA file and visualize the automata using spot.

```python
import spot
import ltlfkit.ltlf2hoa as ltlf
```

### Example "G a"

Spot translates the LTL expression "G a" into a two state Buechi automaton:

```python
aut_F_a = spot.translate('G a', 'deterministic', 'complete')
print(aut_F_a.to_str('hoa'))
aut_F_a
```

With LTLf, the expression "G a" is translated to a DFA with a transition based accept condition.

```python
hoa_str = ltlf.ltl2hoa('G a')
print(hoa_str)
spot.automaton(hoa_str)
```

### Example "F a"

Spot translates the LTL expression "F a" into a two state Buechi automaton:

```python
aut_F_a = spot.translate('F a', 'deterministic', 'complete')
print(aut_F_a.to_str('hoa'))
aut_F_a
```

In LTLf, the expression "F a" is translated into into a DFA with transition based accept conditions:

```python
hoa_str = ltlf.ltl2hoa('F a')
print(hoa_str)
spot.automaton(hoa_str)
```
